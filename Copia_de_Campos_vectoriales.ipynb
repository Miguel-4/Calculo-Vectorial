# **Campos vectoriales**

Documentación de NumPy: https://numpy.org/

Documentación de Matplotlib: https://matplotlib.org/ 

```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline


x,y = np.meshgrid(np.linspace(-10,10,12),np.linspace(-10,10,12)) #Genera los intervalos en x y y

u = 1 # Primera componente del campo vectorial
v = x  #Segunda componente del campo vectorial

t=np.linspace(-2.7,2.7,20)
x2=2*t
y2=t**2+2
plt.plot(x2,y2)

k=np.linspace(-3,3,20)
x3=2*k
y3=k**2 
plt.plot(x3,y3)
t=np.linspace(-3.5,3.5,20)
x4=2*t
y4=t**2 -3
plt.plot(x4,y4)
t=np.linspace(-3.8,3.8,20)
x5=2*t
y5=t**2 -5
plt.plot(x5,y5)
t=np.linspace(-4,4,20)
x6=2*t
y6=t**2 -7
plt.plot(x6,y6)
t=np.linspace(-4.2,4.2,20)
x7=2*t
y7=t**2 -9  
plt.plot(x7,y7)

plt.quiver(x,y,u,v, color = 'green')
plt.show()
```

```python
x,y = np.meshgrid(np.linspace(-1,1,10),np.linspace(-1,1,15)) #Genera los intervalos en x y y

u = x # Primera componente del campo vectorial
v = y  #Segunda componente del campo vectorial

t = np.linspace(0,1,10)
x_1 = t
y_1 = t**2 

plt.plot(x_1,y_1)
plt.quiver(x,y,u,v, color = 'green')
plt.show()
```

```python
x,y = np.meshgrid(np.linspace(-1.5,1.5,15),np.linspace(-1,1,15)) #Genera los intervalos en x y y

u = x # Primera componente del campo vectorial
v = y  #Segunda componente del campo vectorial

t = np.linspace(0,2*np.pi,40)
x_1 = np.cos(t)
y_1 = np.sin(t) 

plt.plot(x_1,y_1)
plt.quiver(x,y,u,v, color = 'green')
plt.show()
```

```python
from mpl_toolkits.mplot3d import axes3d
%matplotlib inline


fig = plt.figure()
ax = fig.gca(projection='3d')

x, y, z = np.meshgrid(np.linspace(-1, 3, 5), # Rango de valores en el eje x con salto de 0.2
                      np.linspace(-1, 3, 5), # Rango de valores en el eje y con salto de 0.2
                      np.linspace(-1, 3, 5)) # Rango de valores en el eje z con salto de 0.2

u = x+y  #Primera componente del campo vectorial
v = y-z #Segunda componente del campo vectorial
w = z**2 #Tercera componente del campo vectorial

t=np.linspace(0,1,20)
x2=t**2
y2=t**3
z2=t**2
plt.plot(x2,y2,z2)

ax.quiver(x, y, z, u, v, w, length=0.1, color = 'green') #Genera el campo vectorial en en espacio

plt.show() #Grafica el campo vectorial
```

```python
from mpl_toolkits.mplot3d import axes3d
%matplotlib inline


fig = plt.figure()
ax = fig.gca(projection='3d')

x, y, z = np.meshgrid(np.linspace(-1, 3, 5), # Rango de valores en el eje x con salto de 0.2
                      np.linspace(-1, 3, 5), # Rango de valores en el eje y con salto de 0.2
                      np.linspace(-1, 3, 5)) # Rango de valores en el eje z con salto de 0.2

u = np.sin(x)  #Primera componente del campo vectorial
v = np.cos(y) #Segunda componente del campo vectorial
w = x*z #Tercera componente del campo vectorial

t=np.linspace(0,1,20)
x2=3*t**3
y2=-t**2
z2=t
plt.plot(x2,y2,z2)

ax.quiver(x, y, z, u, v, w, length=0.1, color = 'green') #Genera el campo vectorial en en espacio

plt.show() #Grafica el campo vectorial
```

```python
from mpl_toolkits.mplot3d import axes3d


fig = plt.figure()
ax = plt.axes(projection='3d')

x, y, z = np.meshgrid(np.linspace(-1.2, 1.2, 10), # Rango de valores en el eje x divide el intervalo [-2,2] en 10 partes
                      np.linspace(-1.2, 1.2, 10), # Rango de valores en el eje y divide el intervalo [-2,2] en 10 partes
                      np.linspace(0, 13, 10)) # Rango de valores en el eje z divide el intervalo [-2,2] en 10 partes

u = y*z  #Primera componente del campo vectorial
v = x*z #Segunda componente del campo vectorial
w = x*y #Tercera componente del campo vectorial

t = np.linspace(0,4*np.pi,40) 

x_1 = np.cos(t)
y_1 = np.sin(t)
z_1 = t

plt.plot(x_1, y_1, z_1)

ax.quiver(x, y, z, u, v, w, length=0.01, color = 'green') #Genera el campo vectorial en en espacio

plt.show() #Grafica el campo vectorial
```
